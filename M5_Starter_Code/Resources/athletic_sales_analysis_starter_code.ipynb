```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.

athletic_sales_2020 = pd.read_csv('Resources/athletic_sales_2020.csv')
athletic_sales_2021 = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame

athletic_sales_2020
```

```python
# Display the 2021 sales DataFrame

athletic_sales_2021
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.

athletic_sales_2020.dtypes
```

```python
# Check the 2021 sales data types.

athletic_sales_2021.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.

combined_athletic_sales = pd.concat([athletic_sales_2020, athletic_sales_2021],\
                                    axis='rows')
combined_athletic_sales
```

```python
# Check if any values are null.

combined_athletic_sales.isnull().any()
```

```python
# Check the data type of each column

combined_athletic_sales.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype

combined_athletic_sales['invoice_date'] = pd.to_datetime(combined_athletic_sales['invoice_date'])
combined_athletic_sales

```

```python
# Confirm that the "invoice_date" data type has been changed.

combined_athletic_sales.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Using groupby. Show the number products sold for region, state, and city. 
# Rename the sum to "Total_Products_Sold".
Reg_State_City_group = combined_athletic_sales.groupby(\
    by=['region', 'state', 'city'])

Total_Products_Sold = Reg_State_City_group['units_sold'].sum()
Total_Products_Sold.sort_values(ascending=False, inplace=True)
Total_Products_Sold.name = 'Total_Products_Sold'

# Show the top 5 results

Total_Products_Sold.head()

                            
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.

products_sold_pivot = pd.pivot_table(combined_athletic_sales, index=['region', 'state', 'city'], values='units_sold', aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"

products_sold_pivot.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.

top_products_sold_pivot = products_sold_pivot.sort_values(by='Total_Products_Sold', ascending=False).head(5)

top_products_sold_pivot
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.

top_region_sales_grouped = combined_athletic_sales.groupby(['region', 'state', 'city'])['total_sales'].sum()

# Rename the "total_sales" column to "Total Sales"

top_region_sales_grouped.sort_values(ascending=False, inplace=True)
top_region_sales_grouped.name = 'Total Sales'

# Show the top 5 results.

top_region_sales_grouped.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.

top_region_sales_pivot = pd.pivot_table(combined_athletic_sales, index=['region', 'state', 'city'], values='total_sales', aggfunc='sum')

# Rename the "total_sales" column to "Total Sales"

top_region_sales_pivot.rename(columns={'total_sales': 'Total_Sales'}, inplace=True)

# Put the results in dollar format.

top_region_sales_pivot['Total_Sales'] = top_region_sales_pivot['Total_Sales'].map('${:,.2f}'.format)

# Show the top 5 results.

top_region_sales_pivot = top_region_sales_pivot.sort_values(by='Total_Sales', ascending=False).head(5)

top_region_sales_pivot
```

### 4. Determine which Retailer had the Most Sales

```python
# Show the total sales for the products sold for each retailer, region, state, and city.

retailer_sales_pivot = pd.pivot_table(combined_athletic_sales, index=['retailer', 'region', 'state', 'city'], values='total_sales', aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"

retailer_sales_pivot.rename(columns={'total_sales': 'Total_Sales'}, inplace=True)

# Show the top 5 results

retailer_sales_pivot.sort_values(by='Total_Sales', ascending=False).head(5)

retailer_sales_pivot

# Show the top 5 results in dollar format

retailer_sales_pivot.sort_values(by='Total_Sales', ascending=False).head(5).map('${:,.2f}'.format)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.

womens_athletic_sales = combined_athletic_sales[combined_athletic_sales['product'] == "Women's Athletic Footwear"]
womens_athletic_sales

```

#### Using `groupby`

```python

# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

womens_athletic_sales_g = womens_athletic_sales.groupby(['retailer', 'region', 'state', 'city'])

# Show the top 5 results

womens_athletic_sales_g['units_sold'].sum().sort_values(ascending=False).head()




```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.

womens_athletic_sales_pivot = pd.pivot_table(combined_athletic_sales[combined_athletic_sales['product'] == "Women's Athletic Footwear"], 
                                             index=['retailer', 'region', 'state', 'city'], 
                                             values='units_sold', 
                                             aggfunc='sum')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

womens_athletic_sales_pivot.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)

# Show the top 5 results for each retailer, region, state, and city in dollar format.

womens_athletic_sales_pivot.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head(5).map('${:,.2f}'.format)







```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.

WAFS_Day = pd.pivot_table(womens_athletic_sales, index=['invoice_date'], values='total_sales', aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"

WAFS_Day.rename(columns={'total_sales': 'Total_Sales'}, inplace=True)

# Show the table in dollar format and sort in descending order.

WAFS_Day.sort_values(by='Total_Sales', ascending=False).head(20).map('${:,.2f}'.format)

```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_sales = WAFS_Day.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
daily_sales.sort_values(by='Total_Sales', ascending=False, inplace=True)

# Show the top 20 results in dollar format.

daily_sales.head(20).map('${:,.2f}'.format)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.

weekly_sales = WAFS_Day.resample('W').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".

weekly_sales.sort_values(by='Total_Sales', ascending=False, inplace=True)

# Put in dollar format.

weekly_sales['Total_Sales'] = weekly_sales['Total_Sales'].map('${:,.2f}'.format)

weekly_sales.head(20)


```
